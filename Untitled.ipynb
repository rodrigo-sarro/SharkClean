```python
def horariocerto(o):
    o = re.sub('.*[Nn]ight.*', 'Night', o)
    o = re.sub('.*[Dd]usk.*', 'Night', o)
    o = re.sub('.*[Mm]orning.*', 'Morning', o)
    o = re.sub('Evening', 'Afternoon',o)
    o = re.sub('.*[Aa]fternoon.*', 'Afternoon', o)
    o = re.sub('[2][0-3]\w\d+', 'Night', o)
    o = re.sub('[1][9]\w\d+', 'Night', o)
    o = re.sub('[1][2-8]\w\d+' , 'Afternoon', o)
    o = re.sub('[0 ][0-9]\w\d+' , 'Morning', o)
    o = re.sub('[1][0-2]\w\d*' , 'Morning', o)
    o = re.sub('[1][0-2]\w\d*' , 'Morning', o)
    o = re.sub('.*[Aa][. -][mM].*' , 'Morning', o)
    o = re.sub('.*[pP][. -][mM].*' , 'Afternoon', o)
    o = re.sub('.*[Mm]idday.*' , 'Morning', o)
    return o
```

```python
shark_time = shark_time_clean.apply(horariocerto)
```

```python
sharkclean['Time'] = sharkclean['Time'].apply(horariocerto)
```
